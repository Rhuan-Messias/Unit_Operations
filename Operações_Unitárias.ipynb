```python
import numpy as np 
import sympy as sp 

#variáveis
C_d_R_e,p_p,p_f,Fi,mi,v_t,k_1,k_2,n,g,Re,d_p = sp.symbols('C_d_R_e,p_p,p_f,fi,mu,v_t,k_1,k_2,n,g,Re,d_p')
# pp = 5.0 # g/cm³
# pf = 1.0 # g/cm³
fi = 0.68
# mi = 0.0009 # Pa.s 
# vt = 3.1 # cm/s

#Primeiro encontrar Cd/Re
eq_CdRe = sp.Eq(C_d_R_e,4*((p_p-p_f)*mi*g)/(3*(v_t**3)*(p_f**2))) 
solve_CdRe = sp.solve(eq_CdRe.subs({p_p:5.0,
                                    p_f:1,
                                    g:9.81,
                                    mi:0.0009,
                                    v_t:3.1
                                    }))[0]
print('Primeiro encontrar o Cd/Re')
print(f'Cd/Re = {solve_CdRe:.4f}\n')

#Agora encontrar Reynolds
eq_Re = sp.Eq(Re,((24/(k_1*C_d_R_e))**(n/2)+(k_2/C_d_R_e)**(n))**(1/n))
#encontrar k1 e k2 para fórmula acima (as equações são tabeladas)
k1 = 0.843*sp.log(fi/0.065,10)
k2 = 5.31-4.88*fi
Re = sp.solve(eq_Re.subs({k_1:k1,
                          k_2:k2,
                          C_d_R_e:solve_CdRe,
                          n:1.3,
                          }))[0]
print('Segundo encontrar Reynolds')
print(f'Re = {Re:.3f}\n')

#usando o reynolds, encontrar dp 
eq_dp = sp.Eq(d_p,Re*mi/(v_t*p_f))
dp = sp.solve(eq_dp.subs({Re:Re,
                          mi:0.0009,
                          v_t:3.1,
                          p_f:1}))[0]
print('Reposta: Valor de dp')
print(f'dp = {dp:.3f} cm')

```
